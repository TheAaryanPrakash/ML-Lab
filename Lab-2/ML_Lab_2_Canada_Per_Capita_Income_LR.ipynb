## 1️⃣ Using Scikit-Learn

```python

from sklearn.linear_model import LinearRegression

def predict_income_sklearn():
    model = LinearRegression()
    X = df[['year']]
    y = df['per capita income (US$)']
    model.fit(X, y)

    prediction = model.predict([[2020]])
    print("Predicted Per Capita Income for 2020 (Sklearn):", prediction[0])

predict_income_sklearn()
```

## 2️⃣ Without Using ML Modules (Manual Calculation)

```python

def predict_income_manual():
    x = df['year'].values
    y = df['per capita income (US$)'].values

    mean_x = np.mean(x)
    mean_y = np.mean(y)

    numerator = np.sum((x - mean_x) * (y - mean_y))
    denominator = np.sum((x - mean_x)**2)

    b1 = numerator / denominator
    b0 = mean_y - b1 * mean_x

    prediction = b0 + b1 * 2020
    print("Predicted Per Capita Income for 2020 (Manual):", prediction)

predict_income_manual()
```
