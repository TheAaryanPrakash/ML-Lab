# Linear Regression – Salary Prediction

This notebook predicts Salary for 12 years of experience using:
1. Scikit-Learn
2. Manual Linear Regression (without ML libraries)

```python

import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/drive/MyDrive/PROJECTS-6/ML-Lab/Lab-2/salary.csv').dropna()
df.head()
```

## 1️⃣ Using Scikit-Learn

```python

from sklearn.linear_model import LinearRegression

def predict_salary_sklearn():
    model = LinearRegression()
    X = df[['YearsExperience']]
    y = df['Salary']
    model.fit(X, y)

    prediction = model.predict([[12]])
    print("Predicted Salary for 12 years experience (Sklearn):", prediction[0])

predict_salary_sklearn()
```

## 2️⃣ Without Using ML Modules (Manual Calculation)

```python

def predict_salary_manual():
    x = df['YearsExperience'].values
    y = df['Salary'].values

    mean_x = np.mean(x)
    mean_y = np.mean(y)

    numerator = np.sum((x - mean_x) * (y - mean_y))
    denominator = np.sum((x - mean_x)**2)

    b1 = numerator / denominator
    b0 = mean_y - b1 * mean_x

    prediction = b0 + b1 * 12
    print("Predicted Salary for 12 years experience (Manual):", prediction)

predict_salary_manual()
```
